```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSVs into DataFrames
brand_df = pd.read_csv("brand_category.csv")
category_df = pd.read_csv("categories.csv")
offer_df = pd.read_csv("offer_retailer.csv")

# brandCategory.csv analysis

# Top 10 brands by receipts  
top_brands = brand_df.sort_values('RECEIPTS', ascending=False).head(10)
print(top_brands)

# Visualize top brand receipts
ax = top_brands['RECEIPTS'].plot.bar(title='Top 10 Brands by Receipts')
ax.set_ylabel('Receipts')
plt.tight_layout()
plt.show()

# Categories for top brands 
print(top_brands['BRAND_BELONGS_TO_CATEGORY'].value_counts())

# categories.csv analysis

# Count child categories per parent
parent_counts = category_df.groupby('IS_CHILD_CATEGORY_TO').size()
print(parent_counts)

# Plot hierarchy 
parent_counts.plot.bar(title='Child Categories per Parent')
plt.tight_layout()
plt.show()

# offerRetailer.csv analysis

# Count offers by retailer  
print(offer_df['RETAILER'].value_counts())

# Count offers by brand
print(offer_df['BRAND'].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = brand_df
df = df.drop('RECEIPTS', axis=1)

# Basic stats
print(df.describe()) 

# Category counts
cat_counts = df['BRAND_BELONGS_TO_CATEGORY'].value_counts()
print(cat_counts)

# Plot counts
cat_counts.nlargest(10).plot.bar()
plt.title('Top 10 Brand Counts per Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Brand counts
print(df['BRAND'].value_counts())

# Unique brand and categories
print("Unique Brands:", df['BRAND'].nunique())
print("Unique Categories:", df['BRAND_BELONGS_TO_CATEGORY'].nunique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = offer_df

# Offer count
print(df['OFFER'].count())

# Unique brand count
print("Unique Brands:", df['BRAND'].nunique())

# Unique retailer count
print("Unique Retailers:", df['RETAILER'].nunique()) 

# Brand counts
brand_counts = df['BRAND'].value_counts()
print(brand_counts.head())

# Plot brand counts
brand_counts.head(10).plot.bar()
plt.title('Top Brand Counts')
plt.ylabel('Number of Offers')
plt.tight_layout()
plt.show()

# Retailer counts
retailer_counts = df['RETAILER'].value_counts()
print(retailer_counts.head())

# Plot retailer counts
retailer_counts.head(10).plot.bar()
plt.title('Top Retailer Counts')
plt.ylabel('Number of Offers')
plt.tight_layout()
plt.show()
```

```python
df = category_df

# Basic stats
print(df.describe())

# Count child categories per parent
parent_counts = df.groupby('IS_CHILD_CATEGORY_TO').size()
print(parent_counts)

# Plot count
parent_counts.nlargest(10).plot.bar()
plt.title('Child Categories per Parent')
plt.tight_layout()
plt.show()

# Unique parents and categories
print("Unique Parents:", df['IS_CHILD_CATEGORY_TO'].nunique())  
print("Unique Categories:", df['PRODUCT_CATEGORY'].nunique())
```
